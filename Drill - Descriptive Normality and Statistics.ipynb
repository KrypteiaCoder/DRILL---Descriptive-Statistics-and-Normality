```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

NORMAL DISTRIBUTION (DARTH VADER FANS)

```python
like_darthvader=np.random.normal(50,1,1000)
```

```python
plt.hist(like_darthvader, bins=50)
plt.xlabel("Popularity of Darth Vader over time")
plt.ylabel("Number of people")
```

```python
plt.hist(like_darthvader, bins=50)
plt.xlabel("Popularity of Darth Vader over time")
plt.ylabel("Number of people")
# Add a vertical line at the mean.
plt.axvline(like_darthvader.mean(), color='b', linestyle='solid', linewidth=2)
# Add a vertical line at one standard deviation above the mean.
plt.axvline(like_darthvader.mean() + like_darthvader.std(), color='b', linestyle='dashed', linewidth=2)
#Add a vertical line at one standard deviation below the mean.
plt.axvline(like_darthvader.mean() - like_darthvader.std(), color='b', linestyle='dashed', linewidth=2)
# Print the histogram.
plt.show()
```

Evaluate whether the descriptive statistics provided useful information about the variable. Can you identify any common characteristics of the distributions that could be usefully described using the mean and/or standard deviation, versus the ones that could not?
Looks like as time progresses, the populatrity of darth vader increases then fades away with fewer people becoming a fam.  We can infer from the mean and std deviation that people have to warm up to his character, but then with the passage of time, they either forget about him or become not impressed.

POISSON DISTRIBUTION (NUMBER OF VISITS TO TRAUMA CENTER IN A 24 HOUR DAY)

```python
visits=np.random.poisson(8,100)
```

```python
plt.hist(visits, bins=20)
plt.xlabel("24 Hour Day")
plt.ylabel("Visits to the Trauma Center")
# Add a vertical line at the mean.
plt.axvline(visits.mean(), color='b', linestyle='solid', linewidth=2)
# Add a vertical line at one standard deviation above the mean.
plt.axvline(visits.mean() + visits.std(), color='b', linestyle='dashed', linewidth=2)
#Add a vertical line at one standard deviation below the mean.
plt.axvline(visits.mean() - visits.std(), color='b', linestyle='dashed', linewidth=2)
# Print the histogram.
plt.show()
```

Evaluate whether the descriptive statistics provided useful information about the variable. Can you identify any common characteristics of the distributions that could be usefully described using the mean and/or standard deviation, versus the ones that could not?
It would seem that as the day enters the daylight phase and more people are alert and awake, the number of people visiting the trauma center increases.
